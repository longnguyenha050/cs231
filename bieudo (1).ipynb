```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation for metrics comparison
methods = ["ResNet152-Flat", "ResNet152-IVF", "HOG-Flat", "HOG-IVF"]
map_top1 = [0.2365, 0.2173, 0.0251, 0.0081]
map_top5 = [0.4297, 0.3522, 0.0392, 0.0087]
map_top10 = [0.4514, 0.3548, 0.0424, 0.0106]

recall_at1 = [0.2365, 0.2173, 0.0251, 0.0081]
recall_at5 = [0.4492, 0.3687, 0.0562, 0.0093]
recall_at10 = [0.5026, 0.3763, 0.0772, 0.0223]

# Plotting Mean Average Precision
plt.figure(figsize=(10, 6))
x = np.arange(len(methods))
width = 0.2

# Replotting with distinct colors for better readability
# Adjusting mAP plot
plt.figure(figsize=(10, 6))
plt.bar(x - width, map_top1, width, color='skyblue', label='mAP@Top-1')
plt.bar(x, map_top5, width, color='lightgreen', label='mAP@Top-5')
plt.bar(x + width, map_top10, width, color='salmon', label='mAP@Top-10')

plt.xlabel('Methods')
plt.ylabel('Mean Average Precision (mAP)')
plt.title('Comparison of mAP across Methods ')
plt.xticks(x, methods, rotation=15)
plt.legend()
plt.tight_layout()
plt.show()

# Replotting Recall with clearer separation using distinct markers and solid lines for all
plt.figure(figsize=(10, 6))
x = np.arange(len(methods))
width = 0.2

plt.bar(x - width, recall_at1, width, color='skyblue', label='Recall@1')
plt.bar(x, recall_at5, width, color='lightgreen', label='Recall@5')
plt.bar(x + width, recall_at10, width, color='salmon', label='Recall@10')

plt.xlabel('Methods')
plt.ylabel('Recall')
plt.title('Comparison of Recall Across Methods')
plt.xticks(x, methods, rotation=15)
plt.legend()
plt.tight_layout()
plt.show()

```
